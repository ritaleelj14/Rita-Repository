### Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for plots
sns.set_style("whitegrid")
```

#### Distributions of Key Numerical Features

```python
# Distribution of Price_SGD
plt.figure(figsize=(10, 6))
sns.histplot(df['Price_SGD'], kde=True, bins=30)
plt.title('Distribution of Laptop Prices (SGD)')
plt.xlabel('Price (SGD)')
plt.ylabel('Frequency')
plt.show()

# Distribution of RAM_GB
plt.figure(figsize=(10, 6))
sns.histplot(df['RAM_GB'], kde=True, bins=20)
plt.title('Distribution of RAM (GB)')
plt.xlabel('RAM (GB)')
plt.ylabel('Frequency')
plt.show()

# Distribution of Storage_GB
plt.figure(figsize=(10, 6))
sns.histplot(df['Storage_GB'], kde=True, bins=20)
plt.title('Distribution of Storage (GB)')
plt.xlabel('Storage (GB)')
plt.ylabel('Frequency')
plt.show()
```

#### Relationships between Categorical Features and Price

```python
# Box plot of Price by Brand
plt.figure(figsize=(12, 7))
sns.boxplot(x='Brand', y='Price_SGD', data=df)
plt.title('Laptop Price Distribution by Brand')
plt.xlabel('Brand')
plt.ylabel('Price (SGD)')
plt.xticks(rotation=45)
plt.show()

# Box plot of Price by CPU
# Limiting to top N CPUs for better readability, e.g., top 10
top_cpus = df['CPU'].value_counts().nlargest(10).index
df_top_cpus = df[df['CPU'].isin(top_cpus)]

plt.figure(figsize=(14, 7))
sns.boxplot(x='CPU', y='Price_SGD', data=df_top_cpus)
plt.title('Laptop Price Distribution by Top 10 CPUs')
plt.xlabel('CPU')
plt.ylabel('Price (SGD)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Data Quality Improvement Suggestions for an Analysable Dataset

Based on the current EDA, here are some suggestions to improve data quality for better analysis:

1.  **Standardize Categorical Data:**
    *   **CPU/GPU/Model Naming:** Ensure consistent naming conventions. For example, 'Intel i5-14600H' and 'i5-14600H' should be treated as the same. This might involve text cleaning (e.g., removing extra spaces, standardizing case) or mapping variations to a single representative value. If there are many distinct entries, consider grouping similar CPUs/GPUs (e.g., by generation or performance tier) to reduce cardinality and simplify analysis.
    *   **Brand:** While current brands seem clean, in larger datasets, typos or alternative spellings (e.g., 'HP' vs. 'Hewlett-Packard') might occur and require standardization.

2.  **Feature Engineering:**
    *   **Price Normalization/Scaling:** For modeling purposes, `Price_SGD` might benefit from normalization or scaling, especially if it's highly skewed, to ensure features contribute equally to model training.
    *   **Combine Discounts:** If `Brand_Discount` and `Member_Discount` are related to `Discount_percent`, consider creating a single 'Total Discount' feature or analyzing their interaction. If `Brand_Discount` and `Member_Discount` are already implicitly included in `Price_SGD` after applying `Discount_percent`, clarify their role or remove redundant features.
    *   **Performance Tiers:** Create new categorical features for `RAM_GB`, `Storage_GB`, `CPU`, and `GPU` (e.g., 'Entry-level', 'Mid-range', 'High-end') to simplify analysis and comparison across different specifications.
    *   **Price per Feature:** For example, 'Price per GB RAM' or 'Price per inch of screen' could provide normalized metrics for comparison.

3.  **Outlier Analysis:**
    *   While not immediately apparent from the descriptive statistics, visualizations (especially box plots) might reveal outliers in numerical columns like `Price_SGD`, `RAM_GB`, `Storage_GB`, and `Weight_kg`. Investigate these outliers to determine if they are data entry errors or genuinely extreme values that need special handling (e.g., capping, transformation, or removal).

4.  **Consistency Checks:**
    *   **Logical Constraints:** For instance, ensure that `Discount_percent` is always between 0 and 100. (The current data appears to be fine in this regard).

5.  **Consider adding more features:**
    *   If available, adding features like 'release year', 'customer ratings', 'battery life', or 'display resolution' could enrich the dataset for a more comprehensive analysis and potentially improve predictive models.
